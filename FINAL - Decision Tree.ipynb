```python
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt  # Import matplotlib

# Read your dataset (adjust the file_path accordingly)
file_path = 'kaggle_converted_dates.csv'
data = pd.read_csv(file_path)

# Assuming 'work_life_balance', 'culture_values', 'career_opp', 'comp_benefits' are your feature columns
features = ['overall_rating','work_life_balance', 'culture_values', 'career_opp', 'comp_benefits']
X = data[features][:10]  # Select the first 100 rows

# Assuming your target column is named 'target'
y = data['recommend'][:10]  # Select the first 100 rows

# Fit the decision tree model
dtree = tree.DecisionTreeClassifier()
dtree.fit(X, y)

# Visualize the decision tree
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
tree.plot_tree(dtree, feature_names=features, class_names=['x', 'v', 'o'])
plt.show()

```

```python
#filtered out no opinion 
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt

# Read your dataset (adjust the file_path accordingly)
file_path = 'kaggle_converted_dates.csv'
data = pd.read_csv(file_path)

# Assuming 'work_life_balance', 'culture_values', 'career_opp', 'comp_benefits' are your feature columns
features = ['overall_rating','work_life_balance', 'culture_values', 'career_opp', 'comp_benefits']
X = data[features][:100]  # Select the first 100 rows

# Assuming your target column is named 'target'
y = data['recommend'][:100]  # Select the first 100 rows

# Specify the classes to be displayed
class_names = ['v', 'r', 'x']

# Fit the decision tree model
dtree = tree.DecisionTreeClassifier()
dtree.fit(X, y)

# Visualize the decision tree with specific classes
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
tree.plot_tree(dtree, feature_names=features, class_names=class_names, filled=True)
plt.show()
```
